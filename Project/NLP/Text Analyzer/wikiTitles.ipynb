```python
import pandas as pd
import seaborn as sns
from bs4 import BeautifulSoup
import requests
```

```python
titles = pd.read_csv("title.csv")
titles
```

```python
titleText = []

for i in range(100):

    link = "http://en.wikipedia.org/wiki/Special:Random"


    linkResponse = requests.get(link, verify = False)
    content = BeautifulSoup(linkResponse.content, "html.parser")

    for tag in content.find("title"):
        title = tag.text.strip()
        titleText.append(title)
```

```python
def titleCleaner(input):
    output = ""
    for i in range(0, len(input) - 12):
        output += input[i]
        
    return output
```

```python
titleWiki = pd.DataFrame(titleText, columns = ["title"])
titles = pd.concat([titles, titleWiki], axis = 0).reset_index(drop = True)
```

```python
titles
```

```python
titles["title"] = titles["title"].apply(titleCleaner)
titles
```

```python
titles.to_csv("title.csv", index = False)
```
