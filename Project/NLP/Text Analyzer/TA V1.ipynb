```python
'''
Phonetic Codex
remove vowels first, 
retain first letter of the words
b, f, p, v → 1
c, g, j, k, q, s, x, z → 2
d, t → 3
l → 4
m, n → 5
r → 6
h, w, y→ unencoded/removed
'''
```

```python
def soundex(input, aschEncoded):
    
    output = []
    input = input.lower()
    
    temp = [c for c in input if c not in ["a", "e", "i", "o", "u"]]
    
    temp = temp[1:len(temp)]
    for i in temp:
        
        if(i in ["b", "f", "p", "v"]):
            output.append("1")
            
        elif(i in ["c", "g", "j", "k", "q", "s", "x", "z"]):
            output.append("2")
            
        elif(i in ["d", "t"]):
            output.append("3")
        
        elif(i in ["l"]):
            output.append("4")
            
        elif(i in ["m", "n"]):
            output.append("5")
            
        elif(i in ["r"]):
            output.append("6")
            
        elif(i in ["h", "w", "y"]):
            #asch is french phoneme for h
            if(aschEncoded):
                output.append(i.upper()) 
        else:
            output.append("")
    return output
```
