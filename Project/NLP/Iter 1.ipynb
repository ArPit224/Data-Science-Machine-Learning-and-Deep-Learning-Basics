```python
import pandas as pd
import numpy  as np
import string 
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.corpus import stopwords
import urllib
from bs4 import BeautifulSoup
import requests
from nltk.corpus import words
```

```python
def textCleaner(input):
    input = input.lower()
    output = "".join([c for c in input if c not in string.punctuation])
    output = [c for c in output.split(" ") if c not in stopwords.words("english")]
    
    return output
```

```python
contentText = []
titleText = []

for i in [1]:

    #link = "http://en.wikipedia.org/wiki/Special:Random"
    #link = "https://en.wikipedia.org/wiki/English_Wikipedia"
    link = "https://en.wikipedia.org/wiki/Jason_Sendwe"

    linkResponse = requests.get(link, verify = False)
    content = BeautifulSoup(linkResponse.content, "html.parser")


    for tag in content.find_all("p")[1:]:
        text = tag.text.strip()
        contentText.append(text)
    
    for tag in content.find("title"):
        title = tag.text.strip()
        titleText.append(title)
    
    
```
