```python
from random import randint
```

```python
def LotStoragePro(input):
    
    if(input >= 0):
        return input
    
    else:
        
        seed = randint(-1, 1)
        value = df["LotFrontage"].mean() +  (seed * df["LotFrontage"].std())
        
        return value
```

```python
df["LotFrontage"] = df["LotFrontage"].apply(LotStoragePro)
```

```python
error = 0

for i in range(0, 1460):
    if(df["LotFrontage"][i] != dataMainTrain["LotFrontage"][i]):
        error += 1
    
error
    
```

```python
print(df["LotFrontage"].std(), dataMainTrain["LotFrontage"].std())
```

```python
while(True):
    df["LotFrontage"] = dataMainTrain["LotFrontage"]
    
    df["LotFrontage"] = df["LotFrontage"].apply(LotStoragePro)
    
    if(df["LotFrontage"].mean() >= 0.99 * dataMainTrain["LotFrontage"].mean()):
        print(df["LotFrontage"].mean(), dataMainTrain["LotFrontage"].mean())
        break
```
