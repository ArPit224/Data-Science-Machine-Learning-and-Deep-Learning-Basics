```python
def powell(input):
    
    try: 
        open(input, "r")
    
    except:
        return pd.concat([df, gentle])
        
    return pd.concat([df, pd.read_csv(input)])
   
```

```python
for i in [2021, 2023]:
    for j in range(1, 9):
        for k in range(1, 32):
            
            if(k < 10):
                input = "Tägliche Aktivitätswerte/" + str(i) + "-" + "0" + str(j) + "-" + "0" + str(k) + ".csv"
                
            else:
                input = "Tägliche Aktivitätswerte/" + str(i) + "-" + "0" + str(j) + "-" + str(k) + ".csv"
             
            df = powell(input)
```
