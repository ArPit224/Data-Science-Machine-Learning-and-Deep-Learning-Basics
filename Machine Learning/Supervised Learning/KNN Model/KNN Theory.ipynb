```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Classified Data", index_col = 0)
df = data.copy(deep = True)
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.exceptions import FitFailedWarning


scalar = StandardScaler()
scalar.fit(df.drop("TARGET CLASS", axis = 1))
```

```python
scaledFeatures = scalar.transform(df.drop("TARGET CLASS", axis = 1))
```

```python
scaledFeatures
```

```python
dfFeat = pd.DataFrame(scaledFeatures, columns = df.drop("TARGET CLASS", axis = 1).columns)
```

```python
dfFeat
```

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(dfFeat, df["TARGET CLASS"], test_size = 0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import classification_report, confusion_matrix

```

```python
error = []

for n in range(1, 700):
    KNNModel = KNeighborsClassifier(n_neighbors = n)
    KNNModel.fit(xTrain, yTrain)
    predictions = KNNModel.predict(xTest)
    
    errorRate = np.mean(predictions != yTest)
    
    error.append(errorRate)
    
    
```

```python
error
```

```python
plt.figure(figsize = (19, 10))
sns.lineplot(data = error, markers = True)
```

```python
KNNModel = KNeighborsClassifier(n_neighbors = 28)
KNNModel.fit(xTrain, yTrain)
predictions = KNNModel.predict(xTest)
print(classification_report(yTest, predictions))
```
