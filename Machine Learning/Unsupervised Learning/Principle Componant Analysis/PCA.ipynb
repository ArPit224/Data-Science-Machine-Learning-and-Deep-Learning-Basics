```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer

dbMain = load_breast_cancer()
```

```python
dbMain.keys()
```

```python
dbMainData = pd.DataFrame(data = dbMain["data"], columns = dbMain["feature_names"])
dbMainTarget = pd.DataFrame(data = dbMain["target"], columns = ["target"])
```

```python
db = dbMainData.copy(deep = True)
```

```python
db
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
scaler.fit(db)
```

```python
scaledData = scaler.transform(db)
```

```python
scaledData = pd.DataFrame(data = scaledData, columns = dbMain["feature_names"])
scaledData
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
```

```python
pca.fit(scaledData)
```

```python
xPca = pca.transform(db)
```

```python
xPca
```

```python
dfPCAcomp = pd.DataFrame(data = pca.components_, columns = dbMain["feature_names"])
dfPCAcomp
```

```python


plt.figure(figsize = (19.2, 10.8))
sns.heatmap(dfPCAcomp, cmap = "plasma", )
```
