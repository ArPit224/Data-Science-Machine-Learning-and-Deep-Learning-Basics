```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
```

```python
dataMain = make_blobs(4000, n_features = 2, centers = 4, cluster_std = 2)
```

```python
plt.figure(figsize = (19.2, 10.8))
plt.scatter(dataMain[0][:, 0], dataMain[0][:, 1], c = dataMain[1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4)
```

```python
kmeans.fit(dataMain[0])
print(kmeans.labels_)
```

```python
fig, (ax1, ax2)  = plt.subplots(1, 2, sharey = True, figsize = (19.2, 10.8))
ax1.set_title("Original")
ax1.scatter(dataMain[0][:,0], dataMain[0][:, 1], c = kmeans.labels_)
ax2.set_title("Predicted")
ax2.scatter(dataMain[0][:,0], dataMain[0][:, 1], c = dataMain[1])
```
