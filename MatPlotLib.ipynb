```python
fig = plt.figure(dpi = 1000)

x = np.arange(0, 10, 0.1)
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, 2**(x), label = 'X Mother', color = 'red', linewidth = 20, alpha = 0.1)
axes.plot(x, 2**x, label = 'X Smother', color = 'green', lw = 0.1, linestyle = '--', marker = 'o',
          markersize = 10, markerfacecolor = 'Yellow', markeredgecolor = 'red')

axes.set_xlim([1,9])

axes.legend()
```
