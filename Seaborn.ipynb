```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.describe()
```

```python
sns.displot(tips['total_bill'], kde = True, bins = 244)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde')
```

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

CATPLOT

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.std)
```

```python
sns.countplot(x = 'sex', data = tips)
```

```python
sns.boxplot(x = 'day', y = 'tip', data = tips, hue = 'smoker')
```

```python
sns.boxplot(x = 'sex', y = 'total_bill', data = tips, hue = 'smoker')
```

```python
sns.violinplot(x = 'day', y = 'tip', data = tips, hue = 'smoker', split = True)
```

```python
sns.stripplot(data = tips, x = 'smoker', y = 'tip', hue = 'sex', jitter = True)
```

```python
sns.swarmplot(data = tips,  x = 'sex', y = 'tip')
sns.violinplot(data = tips, x = 'sex', y = 'tip', color = 'black')
```

```python
sns.catplot(data = tips, x = 'day', y = 'tip', kind = 'violin')
```

GRID

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
iris_grid = sns.PairGrid(iris)
iris_grid.map(sns.boxplot)
```

```python
iris_grid = sns.PairGrid(iris)
iris_grid.map_diag(sns.histplot)
iris_grid.map_lower(sns.kdeplot, palette = 'seismic')
iris_grid.map_upper(plt.scatter)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', 
           markers = ['o', 'v'], scatter_kws = {'s': 100})
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips,
           col = 'sex', row = 'day', hue = 'smoker',
           aspect = 0.6, height = 10)
```
